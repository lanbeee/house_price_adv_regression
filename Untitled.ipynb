```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV, cross_val_predict
from sklearn.preprocessing import StandardScaler , Imputer
from sklearn.metrics import mean_absolute_error
from scipy.stats import norm, skew
from scipy import stats
```

```python
trai = pd.read_csv('train.csv')
y_train = trai['SalePrice']
tess = pd.read_csv('test.csv')
trai = trai.drop(['SalePrice'],axis = 1)
comb = [trai,tess]
```

```python
for dset in comb:
    dset['MSSubClass'] = dset['MSSubClass'].astype(str)
```

```python
catvars1 = trai.dtypes[trai.dtypes == "object"].index
diccat = dict()
for i in catvars1:
    feats = np.sort(np.array(list(Counter(trai[i]).keys())))
    diccat[i] = ' '.join(feats)
    print(i,'  :  ', feats)
    print()
a20 = [i for i in catvars1 if max(Counter(trai[i]).values()) > 1438]
a35 = [i for i in catvars1 if max(Counter(trai[i]).values()) > 1413]
for i in a20:
    print(i,Counter(trai[i]))
print()
for i in a35:
    print(i,Counter(trai[i]))
print(catvars1)
```

```python
numeric_feats = trai.dtypes[trai.dtypes != "object"].index
skewed_feats = trai[numeric_feats].apply(lambda x: skew(x.dropna())) 
skewed_feats = skewed_feats[skewed_feats > 0.75]
skewed_feats = skewed_feats.index
```

```python
cola1 = ['GarageQual','GarageCond','FireplaceQu','KitchenQual','HeatingQC','BsmtQual','BsmtCond','ExterCond','ExterQual']
cola2 = ['GarageType','Alley','MasVnrType','MiscFeature','Fence']
for dset in comb:
    dset['LotFrontage'] = dset.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
    dset.drop(a20,axis= 1,inplace= True)
    dset.GarageYrBlt.fillna(0, inplace= True)
    dset[cola1] = dset[cola1].replace({'Ex':5 ,'Gd':4,'TA':3,'Fa':2,'Po' : 1 ,np.NaN : 0})
    dset[cola2] = dset[cola2].fillna('None')
    dset['MasVnrArea'].fillna(0,inplace = True)
    dset['Electrical'].fillna(dset.Electrical.mode()[0],inplace = True)
    dset['TotalSF'] = dset['GrLivArea'] + dset['1stFlrSF'] + dset['TotalBsmtSF']
    dset.drop(['Id'],axis= 1,inplace= True)
    dset['MSSubClass'] = dset.MSSubClass.astype('str')
    dset['LotShape'].replace({'Reg':3,'IR1':2,'IR2':1,'IR3':0},inplace= True)
    dset['LandSlope'].replace({'Gtl':2,'Mod':1,'Sev':0},inplace= True)
    dset['GarageFinish'].replace({'Fin':3,'RFn':2,'Unf':1,np.NaN:0},inplace= True)
    dset['CentralAir'].replace({'N':0,'Y':1},inplace = True)
    dset['BsmtExposure'].replace({'Gd':4, 'Av':3, 'Mn':2, 'No':1, np.NaN:0},inplace = True)
    dset['PavedDrive'].replace({'N':0,'Y':2,'P':1},inplace = True)
    dset['BsmtFinType1'].replace({'GLQ':6,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,np.NaN:0},inplace = True)
    dset['BsmtFinType2'].replace({'GLQ':6 ,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,np.NaN:0},inplace = True)
y_train.drop(trai[trai.TotalSF>10000].index, inplace= True)
trai.drop(trai[trai.TotalSF>10000].index, inplace=True)
trai2 = pd.concat([trai,y_train],axis= 1)
```

```python
catvars = trai.dtypes[trai.dtypes == "object"].index
notcatvar = list(set(trai.columns) - set(catvars))
for dset in comb:
    for cat in catvars:
        dset[cat] = dset[cat].fillna(dset[cat].mode()[0])
    for ncat in notcatvar:
        dset[ncat] = dset[ncat].fillna(dset[ncat].median())
```

```python
for dset in comb:
    dset[skewed_feats] = np.log1p(dset[skewed_feats])
```
